## Seaborn Part 2
The Python visualization library Seaborn is based on
matplotlib and provides a high-level interface for drawing
attractive statistical graphics.

Make use of the following aliases to import the libraries:

```python
from dataidea.packages import plt, sns, np, pd
```

The basic steps to creating plots with Seaborn are

1. Prepare some data
2. Control figure aesthetics
3. Plot with Seaborn
4. Further customize your plot

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style("whitegrid")
```

```python
g = sns.lmplot(x="tip", y="total_bill", data=tips, aspect=2)
g.set_axis_labels("Tip","Total bill(USD)")
plt.show()
```

```python
data = pd.DataFrame({'x':np.arange(1,101),
'y':np.random.normal(0,4,100)})
```

```python
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

```python
fig, ax = plt.subplots()
```

## Plotting with Seaborn
### Axis Grids

```python
# Subplot grid for plotting conditional relationships
g = sns.FacetGrid(titanic, col="survived", row="sex")
g = g.map(plt.hist,"age")
#Draw a categorical plot onto a Facetgrid
```

Subplot grid for plotting pairwise relationships

```python
h = sns.PairGrid(iris)
h = h.map(plt.scatter)
```

```python
sns.pairplot(iris)
plt.show()
```

Grid for bivariate plot with marginal univariate plots

```python
i = sns.JointGrid(x="x",
y="y",
data=data)
i = i.plot(sns.regplot,
sns.distplot)
```

```python
# Plot data and regression model fitsacross a FacetGrid
sns.lmplot(x="sepal_width",
y="sepal_length",
hue="species",
x_ci = 'sd',
data=iris)
plt.show()
```

### Categorical Plots
#### Scatterplot

Scatterplot with one categorical variable

```python
sns.stripplot(x="species",
y="petal_length",
data=iris, hue='species')
plt.show()
```

Categorical scatterplot with non-overlapping points

```python
sns.swarmplot(x="species",
y="petal_length",
data=iris, hue='species')
plt.show()
```

#### Bar Chart
Show point estimates and confidence intervals with scatterplot glyphs

```python
sns.barplot(x="sex",
y="survived",
hue="class",
data=titanic)
plt.show()
```

#### Count Plot

Show count of observations

```python
sns.countplot(
    x="deck", 
    data=titanic, 
    palette="Greens_d", 
    hue='survived')
plt.show()
```

#### Point Plot
Show point estimates and confidence intervals as rectangular bars

```python
sns.pointplot(x="class", y="survived", hue="sex", data=titanic,
            palette={"male":"g","female":"m"}, markers=["^","o"], linestyles=["-","--"])
plt.show()
```

#### Boxplot

```python
sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic)
plt.show()
```

```python
sns.boxplot(data=iris,orient="h")
plt.show()
```

#### Violin Plot

```python
sns.violinplot(x="age",
y="sex",
hue="survived",
data=titanic)
```

### Distribution Plots
Plot univariate distribution

```python
plot = sns.displot(data.y, kde=False, color="b")
```

```python
sns.displot(data.y, kde=True, color="b")
```

### Matrix Plots
Heatmap

```python
# Exclude non-numeric columns from correlation calculation
numeric_cols = tips.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_cols.corr()

# Plotting the heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
